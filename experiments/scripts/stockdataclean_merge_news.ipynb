```python
hdfc_stock_data = pd.read_csv(r'F:\Xai_traderx\data\raw\HDFC_stock_data.csv')
reliance_stock_data = pd.read_csv(r'F:\Xai_traderx\data\raw\Reliance_stock_data.csv')

def clean_stock_data(df):
    
    df = df.drop([0, 1]).reset_index(drop=True)
    
    # Set the correct column names from the third row and reset the index
    df.columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']
    
  
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    

    df = df.dropna(subset=['Date'])
    

    df['Close'] = pd.to_numeric(df['Close'], errors='coerce')
    df['High'] = pd.to_numeric(df['High'], errors='coerce')
    df['Low'] = pd.to_numeric(df['Low'], errors='coerce')
    df['Open'] = pd.to_numeric(df['Open'], errors='coerce')
    df['Volume'] = pd.to_numeric(df['Volume'], errors='coerce')

    return df

df_hdfc = clean_stock_data(hdfc_stock_data)
df_reliance= clean_stock_data(reliance_stock_data)
```

```python
hdfc_news_path = r'F:\Xai_traderx\data\processed\HDFC_news_sentiment_summary.csv'
reliance_news_path = r'F:\Xai_traderx\data\processed\Reliance_news_sentiment_summary.csv'

hdfc_news = pd.read_csv(hdfc_news_path)
rel_news = pd.read_csv(reliance_news_path)
```

```python
hdfc_news.head()
```

```python
hdfc_news['date'] = pd.to_datetime(hdfc_news['date'], errors='coerce').dt.date
rel_news['date'] = pd.to_datetime(rel_news['date'], errors='coerce').dt.date

# Step 2: Convert 'Date' in stock data to datetime.date type and rename to 'date' to match
df_hdfc['Date'] = pd.to_datetime(df_hdfc['Date'], errors='coerce').dt.date
df_reliance['Date'] = pd.to_datetime(df_reliance['Date'], errors='coerce').dt.date

df_hdfc = df_hdfc.rename(columns={'Date': 'date'})
df_reliance = df_reliance.rename(columns={'Date': 'date'})

merged_hdfc = pd.merge(df_hdfc, hdfc_news, on='date', how='left')
merged_reliance = pd.merge(df_reliance, rel_news, on='date', how='left')
```
